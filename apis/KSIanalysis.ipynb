```python
#Merge two data frame to get one table
merger_data= pd.merge(Fatal,Non_Fatal,on='year',how='outer')
merger_data
merger_drop=merger_data.drop(['acclass_x','acclass_y'],axis=1)
merged_new= merger_drop.rename(columns={'accnum_x':'Fatal Accidents', 'accnum_y':'Non-Fatal Accidents'})

merged_final=merged_new.set_index('year')
merged_final
merged_new
```

```python
# Plot stacked bar to include fatal and non fatal attributes.
plt.clf()
Fatal= merged_new['Fatal Accidents']
Non_Fatal= merged_new['Non-Fatal Accidents']
fig=plt.figure(facecolor='white')
ax= fig.add_subplot(1,1,1)
bar_width= 0.75
bar_l=merged_new['year']
tick_pos=[i for i in bar_l]
# Constructing bar graph with fatal and nonfatal data
ax1=ax.bar(bar_l,Fatal, width= bar_width, label ='Fatal',color='r')
ax2 = ax.bar(bar_l, Non_Fatal, bottom=Fatal, width=bar_width, label="Non_Fatal", color="blue")

ax.set_ylabel("Number of Accidents",fontsize=10)
ax.set_xlabel("Year",fontsize=10)
ax.legend(loc="best")
ax.set_title("Timeline for Fatal Vs Non-Fatal Accidents in Toronto")
plt.xticks(tick_pos, ["2007", "2008", "2009", "2010","2011","2012","2013","2014","2015","2016","2017"],fontsize=9)
plt.yticks(fontsize=9)

for r1, r2 in zip(ax1,ax2):
    h1 = r1.get_height()
    h2 = r2.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="bottom",fontsize=8, color="white")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="bottom",fontsize=8, color="white")
plt.show()
plt.savefig('Fatal Vs Non-Fatal')

print()
```
